```python
import numpy as np
import pandas as pd
from ggplot import *
%matplotlib inline
```

```python
data = pd.DataFrame(np.random.rand(200, 3), columns=['x', 'y', 'z'])
p = ggplot(aes(x='x', y='y'), data)
p
```

```python
p + geom_point()
```

```python
p + geom_point() + geom_line()
```

```python
p + geom_point() + geom_line() + stat_smooth(color='blue')
```

```python
p + geom_line() + geom_now_its_art()
```
